```python
from datasets import load_dataset
import pandas as pd
import numpy as np
```

Using load_dataset to directly load the data

```python
dataset = load_dataset('MongoDB/whatscooking.restaurants')
dataset
```

Converting the dataset into a pandas dataframe for ease of use and dropping columns that are required in our application

```python
df = pd.DataFrame(dataset['train'])
df = df.drop(['OutdoorSeating', 'sponsored', 'TakeOut', 'HappyHour', 'embedding', 'DogsAllowed'], axis = 1)
df.head(-1)
```

Dropping rows that have null values in them

```python
df = df.dropna(subset = ['PriceRange','menu'], axis = 0)
df.shape
```

Converting the above data into a csv for futher eda analysis

```python
df.to_csv('data_processed.csv', index=False)
```
